```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the file

df = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
df
```

```python
#Showing DataFrame info.

df.info()
```

```python
#Describing DataFrame

df.describe()
```

```python
#Show a plot of the total number of responses.

df['Response'].hist()
plt.show()
```

```python
#Show a plot of the response rate by the sales channel.

sns.histplot(data=df, x='Sales Channel', hue='Response', stat="percent") 
```

```python
#Show a plot of the response rate by the total claim amount.

sns.displot(
    data=df, x="Total Claim Amount", hue="Response",
    kind="hist", stat="percent", element="poly", fill=False
)

```

```python
#Show a plot of the response rate by income

sns.histplot(data=df, x='Income', hue='Response', stat= "percent" )
```
