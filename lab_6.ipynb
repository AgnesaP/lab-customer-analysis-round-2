```python
#importing the needed libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#Reading the file

df = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
df
```

```python
#Dealing with the data


```

```python
#Showing the dataframe shape

print(df.shape)
```

```python
# Standardizing header names.
df.columns=df.columns.str.replace(' ','_') 
df.columns = df.columns.str.lower()
df = df.rename(columns={'employmentstatus':'employment_status'})
df

```

```python
#The columns that are numerical:

df.select_dtypes(include=np.number).columns.tolist()
```

```python
#The columns that are categorical:

df.select_dtypes(include=object).columns.tolist()
```

```python
#Checking the NaN values

count = df.isna().sum()
count
```

```python
#Replacing the NaN values with "0" for the numeric columns

numeric_columns = df.select_dtypes(include=['number']).columns

# fill -1 to all NaN 
df[numeric_columns] = df[numeric_columns].fillna(0)
```

```python
#Checking for `NaN` values after replacing the numerical with "0"
df.isna().sum()
```

```python
#Storing the months from the dataset in a separate column.

df['month_effective_to_date'] =  pd.to_datetime(df['effective_to_date']).dt.strftime('%B')
df  
```

```python
#Converting dates in datetime format

df['effective_to_date'] = pd.to_datetime(df['effective_to_date'], errors='coerce') 
df['effective_to_date']
```

```python
#Datetime format -Filtering the data to show only the information for the first quarter , ie. January, February and March.

      
filtered_df = df.loc[ df['effective_to_date'].dt.quarter.isin([1])]

filtered_df
```
