### 3.5. Drop outliers on the dataset after investigating boxplots.  

```python
# to check outliers we will plot boxplot. 
fig, axs = plt.subplots(3, 2, figsize=(20,15))

axs[0, 0].boxplot(df['Weight'])
axs[0, 1].boxplot(df['Length1'])
axs[1, 0].boxplot(df['Length2'])
axs[1, 1].boxplot(df['Length3'])
axs[2, 0].boxplot(df['Height'])
axs[2, 1].boxplot(df['Width'])

axs[0, 0].title.set_text('Weight')
axs[0, 1].title.set_text('Length1')
axs[1, 0].title.set_text('Length2')
axs[1, 1].title.set_text('Length3')
axs[2, 0].title.set_text('Height')
axs[2, 1].title.set_text('Width')

plt.show()
```

Box plots shows that there are outliers in weight, length1, length2 and length3. We will now remove these outliers by using interquartile range.

```python
df.describe()
```

#### Removing Outliers

```python
cols = ['Weight','Length1','Length2','Length3'] # one or more

Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

## 4.Data preprocessing

### 4.1. Separate target feature from the explanatory variables

```python
x = df.iloc[: ,[0,2,3,4,5,6]] #Explanatory variables
y = df.iloc[:, [1]] #Target variable
x
```

```python
#Transformation of categorical variable
df["Species"].unique()
```

```python
x = pd.get_dummies(x, drop_first = True)
x
```

### 4.2. Train test split of order 0.2 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
len(x_train)/len(x)
```

```python
print("no of rows x_train = {}".format(len(x_train)))
print("no of rows x_test = {}".format(len(x_test)))
print("no of rows y_train = {}".format(len(y_train)))
print("no of rows y_test = {}".format(len(y_test)))
```

## 5. Train a linear regression in sklearn 

### 5.0. Standardize the data in the right way 

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
```

```python
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

```python
x_train
```

```python
x_test
```

### 5.1. Fit the model to train data 

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.coef_
```

```python
Recall the formula of the multiple linear regression: :  
y = regressor.coef_[0] * X1 + regressor.coef_[1]* X2 + regressor.coef_[2]* X3 ...

weight = -150.92927353Length1 + 497.47711024Length2 - 205.2498758Length3 + 207.05091841Height + 70.99661659Width + 21.19517324Species_Parkki + 84.36643852Species_Perch + 42.38478831Species_Pike + 53.45715531Species_Roach + 132.49307673Species_Smelt + 25.81456146Species_Whitefish  
```

## 6. Evaluate the model 

### 6.1. R2 score on train data 
### 6.2. R2 score on test data

```python
print("Score of Train : {}\n Score of Test : {}".format(regressor.score(x_train,y_train),regressor.score(x_test,y_test)))
```
