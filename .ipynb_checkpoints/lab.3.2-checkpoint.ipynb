```python
# Bonus - Puting all the previously mentioned data transformations into a function.


def f(file_location, date_field, quarter):

    df = pd.read_csv(file_location)
    df

    # 1.Showing the dataframe shape.

    answer1 = ' 1.Showing the dataframe shape', df.shape

    # 2.Standardizing header names
    df.columns=df.columns.str.replace(' ','_') 
    df.columns = df.columns.str.lower()
    df = df.rename(columns={'employmentstatus':'employment_status'})
    answer2 = '2.Standardizing header names',df

    # 3.The columns that are numerical:

    answer3 = '3.The columns that are numerical:',df.select_dtypes(include=np.number).columns.tolist()

    # 4.The columns that are categorical:

    answer4 = '4.The columns that are categorical:',df.select_dtypes(include=object).columns.tolist()
           

 
    # 5.Checking the NaN values
    count = df.isna().sum()
    count

    # 5.Replacing the NaN values with "0"
    df = df.fillna(0)

    # 5.Checking for `NaN` values after replacing them with "0"
    df.isna().sum()

    
    answer5 =   '5. Checking for `NaN` values after replacing them with "0"',  df.isna().sum()
   

    # 6.storing the months from the dataset in a separate column.
    df['new_effective_to_date'] = df[date_field]
    df

    # 6.converting dates in datetime format

    df['new_effective_to_date'] = pd.to_datetime(df['new_effective_to_date'], errors='coerce') 
    df['new_effective_to_date']

    # 6.Datetime format - Extracting the months from the dataset. Then filtering the data to show only the information for the first quarter , ie. January, February and March.

    df['new_effective_to_date'] = pd.to_datetime(df['new_effective_to_date'], format='%Y-%m-%d')

    filtered_df = df.loc[ df['new_effective_to_date'].dt.quarter.isin([quarter])]
    
    answer6 = '6. Extracting the months from the dataset. Then filtering the data to show only quarter ' + str(quarter)+ ': ',filtered_df

    return answer1, answer2, answer3, answer4, answer5, answer6, 'data frame: ', df




f('files_for_lab/csv_files/marketing_customer_analysis.csv', 'effective_to_date',1)     
```
